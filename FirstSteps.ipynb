```python
from langchain.document_loaders.csv_loader import CSVLoader

loader = CSVLoader(file_path=r"data/FundRanking.csv")   # all defaults
rows   = loader.load()

csv_path = r"data/FundRanking.csv"           # adjust if needed
df = pd.read_csv(csv_path)

import os, pandas as pd
from dotenv import load_dotenv
from langchain_openai import AzureOpenAIEmbeddings, AzureChatOpenAI
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA

# ---------- env / creds ----------
load_dotenv(".env", override=True)

embeddings = AzureOpenAIEmbeddings(
    model          = os.getenv("AZURE_OPENAI_EMBEDDING_DEPLOYMENT"),
    azure_endpoint = os.getenv("AZURE_OPENAI_EMBEDDING_ENDPOINT"),
    api_version    = os.getenv("AZURE_OPENAI_EMBEDDING_API_VERSION"),
    api_key        = os.getenv("AZURE_OPENAI_EMBEDDING_API_KEY"),
)
llm = AzureChatOpenAI(
    azure_deployment = os.getenv("AZURE_OPENAI_CHAT_DEPLOYMENT_NAME"),
    api_version      = os.getenv("AZURE_OPENAI_API_VERSION"),
    temperature      = 0,
)

# ---------- load & clean CSV ----------
df = pd.read_csv(r"data/FundRanking.csv")
df.columns = df.columns.str.strip()
df = df.rename(columns={"Return Volatilty": "Return Volatility"})

pct_cols = ["Total Return 3 Year - Ann", "Risk Free Rate (Average)", "Return Volatility"]
for c in pct_cols:
    df[c] = pd.to_numeric(df[c].str.rstrip("%"), errors="coerce") / 100

df["Sharpe_3Y"] = (
    (df["Total Return 3 Year - Ann"] - df["Risk Free Rate (Average)"])
    / df["Return Volatility"]
).round(3)

# quick sanity check without tabulate/markdown
top5 = df.sort_values("Sharpe_3Y", ascending=False)[["Ticker", "Sharpe_3Y"]].head(5)
print("\n--- Top5 by Sharpe_3Y (plain text) ---")
print(top5.to_string(index=False))

# ---------- build retriever ----------
docs = [
    Document(page_content=f"{r.Ticker} has Sharpe_3Y {r.Sharpe_3Y}",
             metadata={"Ticker": r.Ticker, "Sharpe_3Y": r.Sharpe_3Y})
    for _, r in df.iterrows()
]
db        = FAISS.from_documents(docs, embeddings)
retriever = db.as_retriever(search_kwargs={"k": 30})

prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=(
        "You are a fund analyst.\n\n"
        "Context:\n{context}\n\n"
        "Question: {question}\n\n"
        "Answer in a 5row table (Ticker | Return Sharpe Ratio)."
    ),
)
qa = RetrievalQA.from_chain_type(
    llm       = llm,
    chain_type= "stuff",
    retriever = retriever,
    chain_type_kwargs={"prompt": prompt},
)

print("\n--- LLM answer ---")
print(
    qa.invoke(
        {"query": "List the top 5 tickers by Sharpe ratio."}
    )["result"]
)
```

```python
import os
import pandas as pd
import streamlit as st
from dotenv import load_dotenv
from langchain_openai import AzureOpenAIEmbeddings, AzureChatOpenAI
from langchain.vectorstores import FAISS
from langchain.docstore.document import Document
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA

# ------------ 1. initialise once (runs on first load) ------------
@st.cache_resource
def init_pipeline():
    load_dotenv(".env", override=True)

    # 1-a  Embeddings + LLM
    embeddings = AzureOpenAIEmbeddings(
        model          = os.getenv("AZURE_OPENAI_EMBEDDING_DEPLOYMENT"),
        azure_endpoint = os.getenv("AZURE_OPENAI_EMBEDDING_ENDPOINT"),
        api_version    = os.getenv("AZURE_OPENAI_EMBEDDING_API_VERSION"),
        api_key        = os.getenv("AZURE_OPENAI_EMBEDDING_API_KEY"),
    )
    llm = AzureChatOpenAI(
        azure_deployment = os.getenv("AZURE_OPENAI_CHAT_DEPLOYMENT_NAME"),
        api_version      = os.getenv("AZURE_OPENAI_API_VERSION"),
        temperature      = 0,
    )

    # 1-b  CSV  DataFrame  (no heavy cleaning here; reuse your logic if you like)
    df = pd.read_csv("data/FundRanking.csv")
    df.columns = df.columns.str.strip()
    rank_cols = ["Q","U","X","AA","AL","AN","AP"]
    for c in rank_cols:
        if c in df.columns:
            df[c] = pd.to_numeric(df[c], errors="coerce")

    # 1-c  Build documents & vector store
    docs = [
        Document(
            page_content=(
                f"{row['Long Company Name']} ({row['Ticker']}): "
                f"Rank {row['Q']}, U {row['U']}, X {row['X']}, "
                f"AA {row['AA']}, AL {row['AL']}, AN {row['AN']}, AP {row['AP']}"
            ),
            metadata={"Ticker": row["Ticker"], "Rank": row["Q"]},
        )
        for _, row in df.iterrows()
    ]
    store     = FAISS.from_documents(docs, embeddings)
    retriever = store.as_retriever(search_kwargs={"k": 20})

    prompt = PromptTemplate(
        input_variables=["context", "question"],
        template=(
            "You are a fund analyst.\n\n"
            "Context:\n{context}\n\n"
            "Question: {question}\n\n"
            "Return a 10-row table: "
            "Ticker | Long Company Name | Rank(Q) | Rank(U) | Rank(X) | "
            "Rank(AA) | Rank(AL) | Rank(AN) | Rank(AP)"
        ),
    )

    qa = RetrievalQA.from_chain_type(
        llm=llm,
        chain_type="stuff",
        retriever=retriever,
        chain_type_kwargs={"prompt": prompt},
    )
    return df, qa

df, qa = init_pipeline()

# ------------ 2. UI ------------
st.title("Fund-Ranking QA")

with st.expander("Preview of top-10 by overall Rank (Q)"):
    show_cols = ["Long Company Name", "Ticker", "Q","U","X","AA","AL","AN","AP"]
    st.dataframe(df.nsmallest(10, "Q")[show_cols])

query = st.chat_input("Ask about fund rankings")
if query:
    with st.spinner("Crunching..."):
        answer = qa.run(query)
    st.markdown("### Answer")
    st.markdown(answer)
```

```python
"""
Top-10 fund ranking script
--------------------------
 Requires: pandas, python-dotenv, langchain-openai, faiss-cpu
 Put your CSV in data/FundRanking.csv and your API creds in .env
"""

from langchain.document_loaders.csv_loader import CSVLoader
import os, pandas as pd
from dotenv import load_dotenv
from langchain_openai import AzureOpenAIEmbeddings, AzureChatOpenAI  # Azure flavour
# If youre using regular OpenAI, comment the line above and use:
# from langchain.embeddings import OpenAIEmbeddings
# from langchain.chat_models import ChatOpenAI
from langchain.docstore.document import Document
from langchain.vectorstores import FAISS
from langchain.prompts import PromptTemplate
from langchain.chains import RetrievalQA
from tabulate import tabulate

# ---------------------------------------------------------------------
# 0. Helper  map Excel column letters to zero-based index
# ---------------------------------------------------------------------
def excel_col_to_idx(col: str) -> int:
    idx = 0
    for ch in col.upper():
        if not ch.isalpha():
            raise ValueError(f"Bad column letter: {col}")
        idx = idx * 26 + (ord(ch) - ord('A') + 1)
    return idx - 1

# ---------------------------------------------------------------------
# 1. Load CSV (two ways; the CSVLoader step is optional)
# ---------------------------------------------------------------------
loader = CSVLoader(file_path=r"data/FundRanking.csv")   # LangChain rows version
rows   = loader.load()

csv_path = r"data/FundRanking.csv"
df = pd.read_csv(csv_path)

# ---------------------------------------------------------------------
# 2. Basic header clean-up
# ---------------------------------------------------------------------
df.columns = df.columns.str.strip()
df = df.rename(columns={"Return Volatilty": "Return Volatility"})

# ---------------------------------------------------------------------
# 3. Identify the rank columns by Excel position
#    Q  = overall rank
#    U  X  AA  AL  AN  AP  = other ranks
# ---------------------------------------------------------------------
rank_letters          = ["Q", "U", "X", "AA", "AL", "AN", "AP"]
rank_indices          = [excel_col_to_idx(c) for c in rank_letters]
rank_colnames         = [df.columns[i] for i in rank_indices]

overall_rank_col      = rank_colnames[0]       # df.columns corresponding to Q
other_rank_colnames   = rank_colnames[1:]      # the six additional ranks

# Convert rank columns to numeric (in case they came in as strings)
for c in rank_colnames:
    df[c] = pd.to_numeric(df[c], errors="coerce")

# ---------------------------------------------------------------------
# 4. Pull top-10 by overall rank, ascending (rank 1 is best)
# ---------------------------------------------------------------------
top10 = df.nsmallest(10, overall_rank_col)

display_cols = (
    ["Long Company Name", "Ticker", overall_rank_col] + other_rank_colnames
)

print("\n=== Top-10 Funds by Overall Rank (Q) ===")
print(tabulate(top10[display_cols], headers=display_cols,
               tablefmt="pretty", showindex=False))

# ---------------------------------------------------------------------
# 5. Build LangChain Retrieval-QA so you can query in natural language
# ---------------------------------------------------------------------
load_dotenv(".env", override=True)

# ---- Embeddings & LLM (choose Azure or vanilla OpenAI) ----
embeddings = AzureOpenAIEmbeddings(
    model          = os.getenv("AZURE_OPENAI_EMBEDDING_DEPLOYMENT"),
    azure_endpoint = os.getenv("AZURE_OPENAI_EMBEDDING_ENDPOINT"),
    api_version    = os.getenv("AZURE_OPENAI_EMBEDDING_API_VERSION"),
    api_key        = os.getenv("AZURE_OPENAI_EMBEDDING_API_KEY"),
)
llm = AzureChatOpenAI(
    azure_deployment = os.getenv("AZURE_OPENAI_CHAT_DEPLOYMENT_NAME"),
    api_version      = os.getenv("AZURE_OPENAI_API_VERSION"),
    temperature      = 0,
)

# ----- If youre *not* on Azure, uncomment and set OPENAI_API_KEY -----
# embeddings = OpenAIEmbeddings()
# llm = ChatOpenAI(temperature=0)

# ---- Create miniature documents for every row (or just top10) ----
docs = [
    Document(
        page_content=(
            f"{row['Long Company Name']} ({row['Ticker']}): "
            f"Overall Rank {row[overall_rank_col]}, "
            f"U {row[other_rank_colnames[0]]}, "
            f"X {row[other_rank_colnames[1]]}, "
            f"AA {row[other_rank_colnames[2]]}, "
            f"AL {row[other_rank_colnames[3]]}, "
            f"AN {row[other_rank_colnames[4]]}, "
            f"AP {row[other_rank_colnames[5]]}"
        ),
        metadata={
            "Ticker": row["Ticker"],
            "Long Company Name": row["Long Company Name"],
            overall_rank_col: row[overall_rank_col],
        },
    )
    for _, row in df.iterrows()
]

# ---- Vector store & retriever ----
db        = FAISS.from_documents(docs, embeddings)
retriever = db.as_retriever(search_kwargs={"k": 20})

# ---- Prompt template ----
prompt = PromptTemplate(
    input_variables=["context", "question"],
    template=(
        "You are a fund analyst.\n"
        "Use the context to answer the question.\n\n"
        "Context:\n{context}\n\n"
        "Question: {question}\n\n"
        "Answer *only* with a 10-row table:\n"
        "Ticker | Long Company Name | Rank(Q) | Rank(U) | Rank(X) | "
        "Rank(AA) | Rank(AL) | Rank(AN) | Rank(AP).\n"
    ),
)

qa = RetrievalQA.from_chain_type(
    llm       = llm,
    chain_type= "stuff",
    retriever = retriever,
    chain_type_kwargs={"prompt": prompt},
)

# ---- Example interactive query ----
query = "List the top 10 funds by overall rank and include their other ranking columns."
print("\n=== LLM answer ===")
print(qa.invoke({"query": query})["result"])
```
